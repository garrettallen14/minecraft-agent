```python
import pygetwindow as gw
import pyautogui
from PIL import Image
from google.cloud import storage
from dotenv import load_dotenv
import os
import time
```

```python
def take_screenshot(name):
    # Find the window by title
    window_title = 'Minecraft 1.20.2 - Multiplayer (LAN)'
    titles = gw.getAllTitles()
    for title in titles:
        if 'Minecraft' in title:
            window_title = title
            break
    window = gw.getWindowsWithTitle(window_title)[0]  # Adjust index if necessary
    if window:
        screenshot = pyautogui.screenshot(region=(window.left, window.top, window.width, window.height))
        screenshot.save(name)
```

```python
client = storage.Client()
bucket_name = 'vision-bucket-mineflayer-69'

def upload_blob(source_file_name, destination_blob_name):
    bucket = client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)

    blob.upload_from_filename(source_file_name)

    print(f"File {source_file_name} uploaded to {destination_blob_name}.")

def download_blob(source_blob_name, destination_file_name):
    bucket = client.bucket(bucket_name)
    blob = bucket.blob(source_blob_name)

    blob.download_to_filename(destination_file_name)

    print(f"Blob {source_blob_name} downloaded to {destination_file_name}.")

def list_blobs(delimiter=None):
    blobs = client.list_blobs(bucket_name, delimiter=delimiter)

    blob_list = [blob.name for blob in blobs]
    return blob_list

google_credentials = os.getenv('GOOGLE_APPLICATION_CREDENTIALS')
```

```python
time.sleep(5)
num = 0
while True:
    # Take screenshot
    screenshot_name = f'screenshot_{num}.jpg'
    take_screenshot(f'./images/screenshot_{screenshot_name}')

    # Upload to Google Cloud... save as the filename we used (screenshot_num)
    upload_blob(f'./images/screenshot_{screenshot_name}', screenshot_name)
    print('Uploaded screenshot to Google Cloud')
    
    # Check if the filename has been uploaded to the drive
    response_name = f'response_{num}.txt'
    while True:
        if response_name in list_blobs():
            # Download the file
            # download_blob(response_name, f'./responses/{response_name}')
            try:
                os.remove(f'./responses/responses.txt')
            except:
                print('not there yet')

            download_blob(response_name, f'./responses/responses.txt')
            print('Downloaded response from Google Cloud')
            break

        time.sleep(2)
```
